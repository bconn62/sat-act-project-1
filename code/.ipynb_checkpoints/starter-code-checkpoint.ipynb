##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
def get_sd(column):
    std = np.sqrt((1/np.count_nonzero(column))*np.sum([((value-np.mean(column)))**2 for value in column]))
    return std
```

```python
sd = []
sd.append([get_sd(cate) for cate in big_dict.values()])
type(sd)
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

###### Null = The true mean of ACT participation is the same as the true mean of SAT participation
###### Alternative = The means are different

```python
act_part_list = list(superscore.act_participation)
sat_part_list = list(superscore.sat_participation)
participation_t, participation_p = stats.ttest_ind(act_part_list,sat_part_list)
print(np.mean(act_part_list))
print(np.mean(sat_part_list))
print(participation_t, participation_p)
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
sat_part_ci_upper = np.mean(superscore.sat_participation)+(1.96*(get_sd(superscore.sat_participation)/np.sqrt(np.count_nonzero(superscore.sat_participation))))
sat_part_ci_lower = np.mean(superscore.sat_participation)-(1.96*(get_sd(superscore.sat_participation)/np.sqrt(np.count_nonzero(superscore.sat_participation))))
sat_part_ci = [sat_part_ci_lower, sat_part_ci_upper]

act_part_ci_upper = np.mean(superscore.act_participation)+(1.96*(get_sd(superscore.act_participation)/np.sqrt(np.count_nonzero(superscore.act_participation))))
act_part_ci_lower = np.mean(superscore.act_participation)-(1.96*(get_sd(superscore.act_participation)/np.sqrt(np.count_nonzero(superscore.act_participation))))
act_part_ci = [act_part_ci_lower, act_part_ci_upper]
print("SAT participation confidence 95% interval = " + str(sat_part_ci))
print("ACT participation confidence 95% intveral = " + str(act_part_ci))


```

Based on the definition of a confidence interval, these numbers suggest that out of many sampling distributions, 95% of the distributions would contain the true mean of the participation rate within the given intervals. Looking at the data, however, seems to indicate that this is probably not the case. The data is polarized at the ends of the distributions, which means that the data is not normal and therefore that the CI's are probably invalid.
